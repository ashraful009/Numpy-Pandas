# Indexing with Boolean arrays

```python
arr9 = np.random.randint(low=1, high=100, size=20).reshape(4,5)
arr9
```

```python
arr9[arr9>50]
```

```python
arr9[(arr9>50) & (arr9%2!=0)]
```

```python
arr9[(arr9>50) & (arr9%2!=0)]=0
arr9
```
